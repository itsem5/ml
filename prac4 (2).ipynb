```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#file download link:- "https://www.kaggle.com/datasets/shwetabh123/mall-customers" 
#delete this like after letter
df = pd.read_csv(r"C:\Users\ninjaPC\Downloads\archive (2)\ninja mall.csv")
```

```python

x = df.iloc[:,3:]
```

```python
x
```

```python

plt.title('Unclustered Data')
sns.scatterplot(x=x['Annual Income (k$)'],y=x['Spending Score (1-100)'])
```

```python

from sklearn.cluster import KMeans, AgglomerativeClustering
km = KMeans(n_clusters=4)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
km.inertia_
```

```python

sse =[]
for k in range(1,16):
 km = KMeans(n_clusters=k)
 km.fit_predict(x)
 sse.append(km.inertia_)
```

```python
sse
```

```python

plt.title('Elbow Method')
plt.xlabel('Cluster')
plt.ylabel('SSE')
plt.grid()
plt.xticks(range(1,16))
plt.plot(range(1,16),sse,marker='.',color='blue')
```

```python

km=KMeans(n_clusters=5, random_state=0)
```

```python

labels=km.fit_predict(x)
```

```python
 labels
```

```python

cent = km.cluster_centers_
cent
```

```python

plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('Unclustered Data')
plt.xlabel('Annual income')
plt.ylabel('Spending Score')
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'])
plt.subplot(1,2,2)
plt.title('clustered Data')
plt.xlabel('Annual income')
plt.ylabel('Spending Score')
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'], c=labels)
plt.scatter(cent[:,0],cent[:,1],s=100,color='k')
```
